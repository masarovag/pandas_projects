```python
import pandas as pd
from pandas_datareader import data
import matplotlib.pyplot as plt
%matplotlib inline
import datetime as dt
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
emp = pd.read_csv("employees.csv").dropna(how = "any")
emp["Last Login Time"] = pd.to_datetime(emp["Last Login Time"])
emp["Start Date"] = pd.to_datetime(emp["Start Date"])
emp[["Gender", "Team"]] = emp[["Gender", "Team"]].astype("category")
emp["Senior Management"] = emp["Senior Management"].astype("bool")
pd.set_option('display.max_rows', None) 
emp
```

```python
emp.info()
```

```python
team_salary = emp.sort_values(["Team", "Salary"], ascending = [True,False])
pd.set_option('display.max_rows', None) 
team_salary
```

```python
team_salary ["Total_money"] = team_salary["Salary"] + (team_salary["Salary"]/100) * team_salary["Bonus %"]
team_salary = team_salary.sort_values(["Team", "Total_money"])
team_salary = team_salary[["First Name", "Gender", "Start Date","Last Login Time", "Salary", "Bonus %", "Total_money", "Senior Management", "Team"]]
pd.set_option('display.max_rows', None)
team_salary
```

```python
suma = team_salary.pivot_table(values = "Total_money", index = ("Team"), aggfunc = "sum").sort_values(by = "Total_money", ascending = False).astype(int)
suma.rename(columns={'Total_money': 'Sum_of_team'}, inplace=True)
suma
```

```python
mean = team_salary.pivot_table(values = "Total_money", index = ("Team"), aggfunc = "mean").astype(int).sort_values(by = "Total_money", ascending = False)
mean.rename(columns={'Total_money': 'Average per person'}, inplace=True)
mean
```

```python
team_salary.groupby('Team')['Salary'].agg(['mean', 'sum']).sort_values(by = "mean", ascending = False).astype(int).style.format('$ {0:,.0f}')
```

```python
myDate = dt.datetime(2016, 6, 30)
emp["Years_in_work"] = myDate - emp["Start Date"]
emp['Years_in_work'] = emp['Years_in_work'].astype('<m8[Y]')
emp.sort_values(by=["Years_in_work"], inplace = True)
emp = emp[["First Name", "Gender", "Start Date","Last Login Time", "Years_in_work", "Salary", "Bonus %", "Senior Management", "Team"]]
pd.set_option('display.max_rows', None)
emp
```

```python
def rank_performance(Salary):
    if Salary <= 60000:
        return "Poor"
    elif Salary <= 100000:
        return "Fair"
    else:
        return "Great"
```
